```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_csv('Test.csv')
fig=plt.figure()
sns.scatterplot(dataset.index, dataset['Item_MRP'])
plt.show()
```

```python
# Strip Plot
sns.stripplot(x=dataset['Item_Weight'], y=dataset['Item_MRP'])
```

```python
plt.hist(dataset['Item_MRP'])
```

```python
dataset.dtypes
```

```python
# Rug Plot
fig, ax = plt.subplots()
sns.rugplot(dataset['Item_Weight'])
ax.set_xlim(3,9)
plt.show()
```

```python
# Box Plot
plt.boxplot(dataset['Item_Visibility'])
```

```python
# Distplot
sns.set(rc={'figure.figsize':(6,6)})
sns.distplot(dataset['Item_Weight'], color='black',rug=True)
```

```python
# Pie Chart
plt.figure(figsize=(5,5))
plt.pie(dataset['Outlet_Location_Type'].value_counts(),startangle=80,autopct='%.3f',
        labels=['Item_MRP','Item_Weight','Item_Visibility'],shadow=True)
```

```python
# Line Plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_csv('Test.csv')
fig=plt.figure()
sns.set(rc={'figure.figsize':(7,7)})
sns.set(font_scale=1.5)
fig=sns.lineplot(x=dataset.index,y=dataset['Item_Weight'],markevery=1,marker='d',data=dataset,hue=dataset['Item_MRP'])
fig.set(xlabel='Item_Type')
```
