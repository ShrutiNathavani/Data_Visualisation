```python
# Scatterplot
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('Train.csv')
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(dataset['Item_Weight'], dataset['Item_Visibility'], color=['black'])
plt.title('Cities')
plt.xlabel('Types')
plt.ylabel('Frequency')
plt.show()
```
